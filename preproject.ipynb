# Data cleaning and convert Data type


```python
df["release_date"] = pd.to_datetime(df["release_date"], errors='coerce')
df["year"] = pd.to_datetime(df["year"], errors='coerce')
```

```python
def clean_duration(duration):
    if isinstance(duration, str):
        hours = 0
        minutes = 0
        parts = duration.split()
        
        for part in parts:
            if 'h' in part:
                hours = int(part.replace('h', ''))
            elif 'm' in part:
                minutes = int(part.replace('m', ''))
                
        total_minutes = hours * 60 + minutes
        return total_minutes
    return np.nan

df['duration'] = df['duration'].apply(clean_duration)
```

```python
def convert_votes(v):
    if isinstance(v, str):
        v = v.strip().lower()
        if 'm' in v:
            return float(v.replace('m', '')) * 1000000
        elif 'k' in v:
            return float(v.replace('k', '')) * 1000
        elif v.replace('.', '', 1).isdigit():
            return float(v)
    elif isinstance(v, (int, float)):
        return float(v)
    return np.nan

df['votes'] = df['votes'].apply(convert_votes)
```

# detecting outliers

```python
from datasist.structdata import detect_outliers
```

```python
# Only select numeric columns for outlier detection
numeric_cols = df.select_dtypes(include=[np.number]).columns

# Remove rows with outliers in any numeric column
outliers_indices = detect_outliers(df, 0, numeric_cols)
df.drop(index=outliers_indices, inplace=True)

# Replace outliers in each numeric column with the column median
for col in numeric_cols:
    outliers_indices = detect_outliers(df, 0, [col])
    col_median = df[col].median()
    df.loc[outliers_indices, col] = col_median
```
